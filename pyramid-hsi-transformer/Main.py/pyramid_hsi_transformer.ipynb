```python
!pip install spectral
!pip install einops

import scipy.io
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, cohen_kappa_score, classification_report
import seaborn as sns
from torch.utils.data import Dataset, DataLoader
from einops import rearrange
import spectral

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
data = scipy.io.loadmat('/content/KSC.mat')['KSC']
labels = scipy.io.loadmat('/content/KSC_gt.mat')['KSC_gt']

print("Data Shape:", data.shape)
print("Labels Shape:", labels.shape)

# Normalize data
def normalize_data(data):
    data_min = np.min(data)
    data_max = np.max(data)
    norm_data = (data - data_min) / (data_max - data_min)
    return norm_data

data = normalize_data(data)
```

```python
# 3. Create Patches
PATCH_SIZE = 5

def pad_with_zeros(X, margin):
    new_X = np.pad(X, ((margin, margin), (margin, margin), (0, 0)), mode='constant')
    return new_X

def create_patches(X, y, patch_size=PATCH_SIZE):
    margin = patch_size // 2
    X = pad_with_zeros(X, margin)
    patches = []
    labels = []
    for r in range(margin, X.shape[0] - margin):
        for c in range(margin, X.shape[1] - margin):
            if y[r - margin, c - margin] == 0:
                continue
            patch = X[r - margin:r + margin + 1, c - margin:c + margin + 1, :]
            patches.append(patch)
            labels.append(y[r - margin, c - margin])
    return np.array(patches), np.array(labels)

X, y = create_patches(data, labels)
print("Patches shape:", X.shape)
print("Labels shape:", y.shape)

# Split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Create Dataset
class HyperspectralDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        # Ensure labels are within the valid range [0, num_classes - 1]
        self.y = torch.tensor(y, dtype=torch.long) - 1
        self.y = torch.clamp(self.y, 0, len(np.unique(y)) - 2) # Clamp to prevent out-of-bounds errors

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = HyperspectralDataset(X_train, y_train)
test_dataset = HyperspectralDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
# 4. Build the Improved PyFormer Model
class TransformerClassifier(nn.Module):
    def __init__(self, input_dim, num_classes, patch_size=PATCH_SIZE):
        super(TransformerClassifier, self).__init__()
        self.flatten_dim = input_dim * patch_size * patch_size
        self.embedding = nn.Linear(self.flatten_dim, 256)
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=256, nhead=8)
        self.transformer_encoder = nn.TransformerEncoder(self.encoder_layer, num_layers=4)
        self.classifier = nn.Linear(256, num_classes)

    def forward(self, x):
        batch_size = x.shape[0]
        x = x.view(batch_size, -1)
        x = self.embedding(x)
        x = x.unsqueeze(0)  # (1, batch_size, 256)
        x = self.transformer_encoder(x)
        x = x.squeeze(0)
        x = self.classifier(x)
        return x

model = TransformerClassifier(input_dim=data.shape[2], num_classes=len(np.unique(y)) - 1).to(device)
```

```python
# 5. Train the Model
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
num_epochs = 100

train_losses = []
test_losses = []
train_accuracies = []
test_accuracies = []

for epoch in range(num_epochs):
    model.train()
    running_loss = 0
    correct = 0
    total = 0
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += targets.size(0)
        correct += predicted.eq(targets).sum().item()

    train_loss = running_loss / len(train_loader)
    train_acc = correct / total
    train_losses.append(train_loss)
    train_accuracies.append(train_acc)

    # Validation
    model.eval()
    val_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, targets)

            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()

    val_loss = val_loss / len(test_loader)
    val_acc = correct / total
    test_losses.append(val_loss)
    test_accuracies.append(val_acc)

    print(f"Epoch {epoch+1}/{num_epochs}: Train Loss={train_loss:.4f}, Train Acc={train_acc:.4f}, Val Loss={val_loss:.4f}, Val Acc={val_acc:.4f}")
```

```python
# 6. Evaluate the Model
y_true = []
y_pred = []

model.eval()
with torch.no_grad():
    for inputs, targets in test_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        outputs = model(inputs)
        _, predicted = outputs.max(1)
        y_true.extend(targets.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

conf_mat = confusion_matrix(y_true, y_pred)
oa = np.sum(y_true == y_pred) / len(y_true)
aa = np.mean(np.diag(conf_mat) / np.sum(conf_mat, axis=1))
kappa = cohen_kappa_score(y_true, y_pred)

print("Overall Accuracy (OA):", oa)
print("Average Accuracy (AA):", aa)
print("Kappa Coefficient:", kappa)
print("Classification Report:\n", classification_report(y_true, y_pred))

# Plot Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
# 7. Plot Training Curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(test_accuracies, label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()
```

```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# 1. Load Ground Truth
gt_data = sio.loadmat('KSC_gt.mat')
gt = gt_data['KSC_gt']

print("Ground Truth Shape:", gt.shape)

# 2. Flatten Ground Truth for splitting
h, w = gt.shape
gt_flat = gt.flatten()

# 3. Find valid (non-zero) labeled pixels
valid_idx = np.where(gt_flat > 0)[0]

# 4. Split into train, val, test indices
train_idx, temp_idx = train_test_split(valid_idx, test_size=0.6, random_state=42, stratify=gt_flat[valid_idx])
val_idx, test_idx = train_test_split(temp_idx, test_size=0.5, random_state=42, stratify=gt_flat[temp_idx])

# 5. Create empty masks
train_mask = np.zeros_like(gt_flat)
val_mask = np.zeros_like(gt_flat)
test_mask = np.zeros_like(gt_flat)

# 6. Assign labels
train_mask[train_idx] = gt_flat[train_idx]
val_mask[val_idx] = gt_flat[val_idx]
test_mask[test_idx] = gt_flat[test_idx]

# 7. Reshape back to original image size
train_mask = train_mask.reshape(h, w)
val_mask = val_mask.reshape(h, w)
test_mask = test_mask.reshape(h, w)

# 8. Plotting
fig, axs = plt.subplots(1, 4, figsize=(24, 6))

# Full Ground Truth
axs[0].imshow(gt, cmap='jet')
axs[0].set_title('Ground Truth Map (All Labels)')
axs[0].axis('off')

# Train Set
axs[1].imshow(train_mask, cmap='jet')
axs[1].set_title('Training Samples')
axs[1].axis('off')

# Validation Set
axs[2].imshow(val_mask, cmap='jet')
axs[2].set_title('Validation Samples')
axs[2].axis('off')

# Test Set
axs[3].imshow(test_mask, cmap='jet')
axs[3].set_title('Test Samples')
axs[3].axis('off')

plt.tight_layout()
plt.show()
```
